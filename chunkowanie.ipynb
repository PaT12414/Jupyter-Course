```python
def czytaj_linie(filename):

    with open(filename, "r", encoding="utf-8") as f:
    
        for line in f:
    
            yield line.strip()
```

```python
import pandas as pd

def licz_sume_chunks(path):
    suma = 0

    for chunk in pd.read_csv(path, chunksize=50_000):
        suma += chunk["amount"].sum()

    return suma
```
