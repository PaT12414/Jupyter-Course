```python
import numpy as np
import pandas as pd

def process_sales_optimized(path: str) -> pd.DataFrame:
    df = pd.read_csv(path)

    df["country"] = df["country"].astype("category")
    df["product"] = df["product"].astype("category")
    df["amount"] = pd.to_numeric(df["amount"], errors="coerce", downcast="integer")

    mask = (df["amount"] > 1000) & (df["country"] == "PL")
    df["flag"] = mask.astype("int8")

    return df
```

- niepotrzebna kopia DataFrame df.copy która zwiększa zużycie RAM
- użycie df.apply powodowało iterację po wierszach , co jest bardzo wolne w dużych zbiorach danych
- brak optymalizacji typów liczbowych powodował większe zużycie pamięci
