```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "amount": [500, 1500, 2000, 300],
    "country": ["PL", "PL", "DE", "PL"],
    "discount": [0.1, 0.2, 0.0, 0.15]
})
```

```python
mask = (df["country"] == "PL") & (df["amount"] > 1000)
```

```python
df["final_price"] = np.where(
    mask,
    df["amount"] * (1 - df["discount"]),
    df["amount"]
)

df
```
